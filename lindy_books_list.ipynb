```python
from src.logic import *
import matplotlib
import seaborn as sns
```

```python
cols_of_interest = ['title', 'description', 'author', 'rank', 'list_name', 'weeks_on_list']
```

```python
data = get_ny_times_data('auth.yaml', n_batches=None)
```

### Top 10 most frequent weeks on bestseller list by rank

```python
# top 10 weeks on list according to NY-times weeks_of_list field
data_wl = data[data['rank'].notnull()]; data_wl = data_wl[data_wl['weeks_on_list'].notnull()]
data_wl = data_wl.sort_values(by=['rank', 'weeks_on_list'], ascending=[True, False])
data_wl = data_wl[cols_of_interest]
data_wl = data_wl.drop_duplicates(subset=['title', 'author'], keep='first')
wl_top_ten = data_wl.groupby('rank').head(10)
wl_top_ten = wl_top_ten[wl_top_ten['rank'] <= 10.0]
wl_top_ten
```

```python
rank_to_show=1
wl_top_ten[wl_top_ten['rank'] == rank_to_show]
```

### Top 10 most frequent weeks by rank (according to timestamp)

```python
# top 10 weeks on list according to bestseller timestamp
#cols_of_interest.remove('weeks_on_list'); cols_of_interest.append('bestsellers_date')
data_tl = data[data['rank'].notnull()]; data_tl = data_tl[data_tl['bestsellers_date'].notnull()]
data_tl = data_tl[cols_of_interest]
data_tl = data_tl.groupby(['title','description','author','rank','list_name']).agg(['count'])
data_tl = data_tl.sort_index(ascending=False)
data_tl
```
